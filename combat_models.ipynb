```python
#ARM COMBAT MODEL 1
import numpy as np
import matplotlib.pyplot as plt
import math

x=50
y=50
a1=0.4
b1=0.3
h=0.1
n=101
j=0
a=np.zeros([2,n])
time=np.linspace(0,n-1,n)
for i in range(0,n,1):
  f1=a1*y
  f2=b1*x
  a[j][i]=x
  j=j+1
  a[j][i]=y
  x=x+h*f1
  y=y+h*f2
  j=0
#print(a)
plt.plot(time,a[0,:],label='x')
plt.plot(time,a[1,:],label='y')
plt.legend()
plt.grid()
```
